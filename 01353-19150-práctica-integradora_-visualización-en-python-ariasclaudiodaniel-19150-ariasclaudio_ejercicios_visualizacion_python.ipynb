# Realizar estadísticas descriptivas básicas

```python
import numpy as np
import pandas as pd 
# --- Específicas para los gráficos
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
mpl.style.use('bmh') # bmh es el nombre del estilo que vamos a usar
```

```python
salarios = pd.read_csv('HRDataset_v14.csv',index_col=0, parse_dates=True)
salarios.head()
```

```python
print("Estadísticas descriptivas")
salarios.describe()
```

```python
salarios.groupby('RecruitmentSource').Salary.agg(['count','min','max','mean'])
```

# Realizar un histograma con los salarios. qué rango(s) de salarios son los más populares?.

```python
plt.hist(salarios['Salary'],bins= 10, alpha = 1, ec="black")
plt.title("Histograma de salarios")
plt.grid(True)
plt.xlabel('$ Sueldos')
plt.ylabel('Cant')
plt.savefig('Histograma de Sueldos')
plt.show()
```

```python
print(salarios['Salary'].describe())
```

# 3. Realizar un gráfico de violín con los salarios, discriminados por género.

```python
salarios.groupby(['GenderID']).Salary.sum()
```

```python
sal = sns.catplot(x='GenderID', y='Salary',data=salarios ,kind='violin')
sal.set(xlabel="Género", ylabel="Salarios", title="Salarios discriminados por género")
```
