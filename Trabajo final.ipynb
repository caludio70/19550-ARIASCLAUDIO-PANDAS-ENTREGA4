<div class="alert alert-block alert-info">
 💡<b><span style="font-family:Webdings; font-size: 16pt;"> </span></b><b><span style="font-family:Arial">  </span></b>:<b><span style="font-family:courier new"> <b> Grafico histograma con linea de densidad asociada</b></span>  <span style="font-family:Arial"</span>  
<div>   

```python
def univ_hisplot(df,col_name,ax):
    
    plot0=sns.histplot(data = df,x =col_name, ax=ax,color='green', kde=True)
    ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    ax.set_title(f'Distribution of {col_name}',fontdict={'fontsize':8})
    ax.set_xlabel(col_name,fontdict={'fontsize':7})
    ax.set_ylabel(r'Count/Dist.',fontdict={'fontsize':7})
    plt.tight_layout()
```

<div class="alert alert-block alert-info">
 💡<b><span style="font-family:Webdings; font-size: 16pt;">  </span></b><b><span style="font-family:Arial">  </span></b>:<b><span style="font-family:courier new"> <b> Diagrama de cajas</b></span>  <span style="font-family:Arial"</span>  
<div>   

```python
def univ_boxplot(df,col_name,ax):
    
    plot1=sns.boxplot(data = df[df[col_name]!=0][col_name],ax=ax,orient='v')
    ax.set_title('Box plot',fontdict={'fontsize':8})
    ax.set_ylabel(col_name,fontdict={'fontsize':7})
    plt.tight_layout()
```

<div class="alert alert-block alert-info">
    💡 <b><span style="font-family:Webdings; font-size: 16pt;"></span></b><b><span style="font-family:Arial">  </span></b>:<b><span style="font-family:courier new"> <b> Gráfico de Violín</b></span>  <span style="font-family:Arial"</span>  
<div>   

```python
def univ_violinplot(df,col_name,ax):
    
    plot2=sns.violinplot(x = df[df[col_name]!=0][col_name],ax=ax,palette="Set1")
    ax.set_title('Violin Plot',fontdict={'fontsize':8})
    ax.set_xlabel(col_name,fontdict={'fontsize':7})
    plt.tight_layout()
```

<div class="alert alert-block alert-info">
 💡    </span></b><b><span style="font-family:Arial">  </span></b>:<b><span style="font-family:courier new"> <b>Grafica histograma, diagrama de cajas y grafico de violina para colum_name</b></span>  <span style="font-family:Arial"</span>  
<div>   

```python
def plots_univariate_analysis(dataframe,column_name):
    
    fig,axes = plt.subplots(nrows=1,ncols=3,dpi=120,figsize = (10,5))
    fig.suptitle('Graficos para analisis univariados')
    univ_hisplot(dataframe,column_name,axes[0])
    univ_boxplot(dataframe,column_name,axes[1])
    univ_violinplot(dataframe,column_name,axes[2])
    plt.show()
```

<div class="alert alert-block alert-info">
 💡    </span></b><b><span style="font-family:Arial">  </span></b>:<b><span style="font-family:courier new"> <b>Arma la tabla de distribucion de frecuencias de col_name</b></span>  <span style="font-family:Arial"</span>  
<div>   

```python
def freq_dist(df, col_name):
    
    frec_df = df[col_name].value_counts().reset_index().rename(columns={col_name: 'frec_abs',
                                                                        'index': col_name}).sort_values(col_name)
    frec_df['frec_abs_acum'] = frec_df['frec_abs'].cumsum()
    frec_df['frec_rel_%'] = round(100 * frec_df['frec_abs']/len(df[col_name]), 4)
    frec_df['frec_rel_%_acum'] = frec_df['frec_rel_%'].cumsum()
    display(frec_df)
```

<div class="alert alert-block alert-info">
 💡    </span></b><b><span style="font-family:Arial">  </span></b>:<b><span style="font-family:courier new"> <b>
    Muestra el analisis univariado para column_name :<br>
            ✔️-Descripcion estadistica de colum_name <br>
            ✔️-Histograma, diagrama de cajas, y grafico de violin de column_name <br>
            ✔️-Tabla de distribucion de frecuencias"""</b></span>  <span style="font-family:Arial"</span>  
<div>   

```python
def univariate_analysis(dataframe, column_name):
    
    display(Markdown(f'### Descripcion de la columna {column_name} '))
    
    print(dataframe[column_name].describe())
    plots_univariate_analysis(dataframe,column_name)
    
    display(Markdown('### Distribucion de frequencias '))
    freq_dist(dataframe,column_name)
    
```
