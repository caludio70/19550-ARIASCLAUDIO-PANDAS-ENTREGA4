     <b> <span style="font-family:Arial Black">Primer paso</span></b>: Importo las librerías y escribo la url de la API que nos proporciona el ejercicio. <b>

```python
import requests
import json
import pandas as pd
    
url='https://jsonplaceholder.typicode.com/users'  #url de la api donde esta el json 
```

<div class="alert alert-block alert-info">
     <b> <span style="font-family:Arial Black">Segundo paso</span></b>: hacer el get de la url de la API si la respuesta es "200", dicho número nos indica que la conexión es exitosa. <b>
    El método GET, se utiliza para acceder a los datos del recurso específico en este ejercicio de la API brindada, el objeto de respuesta contiene todos los datos enviados desde el servidor en respuesta de la solicitud GET que ejecutamos 
</div>

```python
respuesta =  requests.get(url) 
print(respuesta)
```

<div class="alert alert-block alert-danger">
     <b> <span style="font-family:Arial Black">Nota</span></b>: en el caso de que el get de la url de la API su respuesta sea > "200", la conexión es deficiente. 
</div>

<div class="alert alert-block alert-info">
     <b><span style="font-family:Arial Black"> Tercer paso</span></b>: observamos la estructura del archivo.
</div>

```python
respuesta.text
```

```python
texto = respuesta.text
json.loads(texto)
```

```python
jsondata = json.loads(texto)
```

<div class="alert alert-block alert-success">
     <b><span style="font-family:Arial Black">Cuarto Paso</span></b>: Construímos un DataFrame utilizando el parámetro <span style="font-family:Arial Black">from_dict.</span>  
<div>

- [ ]  <span style="color:gray">**Escribir un código que lea un JSON desde esta esta web api**</span>
- [X]  **Pasarlo a un DataFrame en pandas**
 

```python
df = pd.DataFrame.from_dict(jsondata)
df.head(10)
```
